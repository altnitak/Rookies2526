```python
import time
import pandas as pd
from nba_api.stats.endpoints import boxscoretraditionalv3

retry_rows = []
still_failed = []

def fetch_teamstats_v3(gid, max_tries=6):
    for attempt in range(1, max_tries + 1):
        try:
            bs = boxscoretraditionalv3.BoxScoreTraditionalV3(game_id=gid, timeout=180)
            df = bs.get_data_frames()[2].copy()  # Team stats (2 Zeilen)
            return df
        except Exception as e:
            wait = min(2, 2 * attempt)  # 15,30,45,60,75,90...
            print(f"retry {attempt}/{max_tries} {gid} -> {type(e).__name__} | sleep {wait}s")
            time.sleep(wait)
    return None

for gid in retry_ids:
    df = fetch_teamstats_v3(gid)
    if df is None:
        still_failed.append(gid)
        continue

    df["gameId"] = df["gameId"].astype(str)
    df["teamId"] = df["teamId"].astype(str)

    m = df.merge(meta2, on=["gameId","teamId"], how="left")

    m["team"] = m["TEAM_ABBREVIATION"]
    is_home = m["MATCHUP"].str.contains(r"vs\.", na=False)
    opp = m["MATCHUP"].str.split().str[-1]
    m["home"] = m["team"].where(is_home, opp)
    m["away"] = opp.where(is_home, m["team"])

    m["win"] = (m["WL"] == "W").astype(int)
    m["date"] = pd.to_datetime(m["GAME_DATE"]).dt.strftime("%d.%m.%Y")

    m["type"] = "regular"
    m["season"] = 2026
    m["gameid"] = m["gameId"]
    m["teamid"] = m["teamId"]

    m = m.rename(columns=rename_v3)

    for c in cols_final:
        if c not in m.columns:
            m[c] = pd.NA

    retry_rows.append(m[cols_final])

    time.sleep(6)  # extra freundlich zur API

retry_df = pd.concat(retry_rows, ignore_index=True) if retry_rows else pd.DataFrame(columns=cols_final)
print("retried rows:", len(retry_df), "still_failed:", len(still_failed))
still_failed
```
